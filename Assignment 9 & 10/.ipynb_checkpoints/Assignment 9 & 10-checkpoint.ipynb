```python
import pandas as pd
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.corr().style.background_gradient(cmap='coolwarm', axis=None).set_precision(2)
```

In general i don't see any results that strike me as too far off of what i expected. Except for the bill_depth and bill_length having a weak correlation with eachother. This is something i didn't expect.

## Strongest positive correlation

```python
penguins.plot(kind="scatter", x="body_mass_g", y="flipper_length_mm")
```

## Strongest negative correlation

```python
penguins.plot(kind="scatter", x="flipper_length_mm", y="bill_depth_mm")
```

## Weakest correlation

```python
penguins.plot(kind="scatter", x="bill_depth_mm", y="bill_length_mm")
```

# Assignment 10

```python
vgSales = pd.read_csv("../Assignment 4/vgsales.csv")
```

```python
vgSales.corr().style.background_gradient(cmap='coolwarm', axis=None).set_precision(2)
```

```python
vgSales.plot(kind="scatter", x="Critic_Score", y="User_Score")
```

```python
vgSales.plot(kind="scatter", x="Critic_Score", y="Total_Shipped")
```

When checking the correlations it's interesting to note that the score a critic gives will be rougly similar to what users would score a game. But what's more interesting is that a Critic's score doesn't seem to affect the sales numbers at all. The correlation is very weak, even if you remove the big outliers it doesn't change much of anything.

```python
vgSales[(vgSales.Total_Shipped < 50)].plot(kind="scatter", x="Critic_Score", y="Total_Shipped")
```

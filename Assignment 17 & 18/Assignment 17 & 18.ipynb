```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
penguins = sns.load_dataset("penguins")
penguins = penguins.dropna()
penguins.head()
```

```python
penguins.corr()
```

```python
penguins_train, penguins_test = train_test_split(penguins, test_size=0.3, random_state=42)
print(penguins_train.shape, penguins_test.shape)
```

```python
from sklearn import tree
import graphviz

def plot_tree_regression(model, features):
    # Generate plot data
    dot_data = tree.export_graphviz(model, out_file=None, 
                          feature_names=features,  
                          filled=True, rounded=True,  
                          special_characters=True)  

    # Turn into graph using graphviz
    graph = graphviz.Source(dot_data)  

    # Write out a pdf
    graph.render("decision_tree")

    # Display in the notebook
    return graph 
```

```python
features= ['flipper_length_mm']
target = ['body_mass_g']
dt_regression = DecisionTreeRegressor(max_depth = 2) # Increase max_depth to see effect in the plot
dt_regression.fit(penguins_train[features], penguins_train[target])
```

```python
plot_tree_regression(dt_regression, features)
```

```python
def calculate_rmse(predictions, actuals):
    if(len(predictions) != len(actuals)):
        raise Exception("The amount of predictions did not equal the amount of actuals")
    
    return (((predictions - actuals) ** 2).sum() / len(actuals)) ** (1/2)

## The same function but using a for-loop instead of a vectorized operation. 
# def calculate_rmse(predictions, actuals):
#    if(len(predictions) != len(actuals)):
#        raise Exception("The amount of predictions did not equal the amount of actuals")
#    
#    diffSquared = 0
#    
#    for prediction_i, actual_i in zip(predictions, actuals):
#        diffSquared += (prediction_i - actual_i)**2
#        
#    return (diffSquared/len(actuals))**(1/2)
```

```python
predictionsOnTrainset = dt_regression.predict(penguins_train[features])
predictionsOnTestset = dt_regression.predict(penguins_test[features])

rmseTrain = calculate_rmse(predictionsOnTrainset, penguins_train.flipper_length_mm)
rmseTest = calculate_rmse(predictionsOnTestset, penguins_test.flipper_length_mm)

print("RMSE on training set " + str(rmseTrain))
print("RMSE on test set " + str(rmseTest))
```

```python
vgSales = pd.read_csv("../Assignment 4/vgsales.csv")
vgSales.isna().sum()
```

```python
vgSales = vgSales.fillna(vgSales.mean())
vgSales = vgSales.fillna("Unknown")
```

```python
vgSales.isna().sum()
```

```python
vgSales.corr()
```

```python
vgSales_train, vgSales_test = train_test_split(vgSales, test_size=0.3, random_state=42)
print(vgSales_train.shape, vgSales_test.shape)
```

```python
features= ['Global_Sales', 'Total_Shipped', 'NA_Sales']
target = ['Critic_Score']
dt_regression = DecisionTreeRegressor(max_depth = 5) # Increase max_depth to see effect in the plot
dt_regression.fit(vgSales_train[features], vgSales_train[target])
```

```python
plot_tree_regression(dt_regression, features)
```

```python
predictionsOnTrainset = dt_regression.predict(vgSales_train[features])
predictionsOnTestset = dt_regression.predict(vgSales_test[features])

rmseTrain = calculate_rmse(predictionsOnTrainset, vgSales_train.Critic_Score)
rmseTest = calculate_rmse(predictionsOnTestset, vgSales_test.Critic_Score)

print("RMSE on training set " + str(rmseTrain))
print("RMSE on test set " + str(rmseTest))
```
